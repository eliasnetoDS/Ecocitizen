#Apresentação dos dados em um mapa

```python
#importando a biblioteca folium, necessária para criação do mapa e visualização dos dados
import folium
from folium.plugins import HeatMap
```

```python
#Criando um mapa vazio (Map_objet) e depois adicionando os pontos de focos de calor, no formato de mapa e calor
map_obj = folium.Map(location = [-5.663417, -60.089654], zoom_start = 5)

locais = focos_amazonia[['lat','lon']].values.tolist()
HeatMap(locais).add_to(map_obj)  #gerando o mapa de calor e adicionando no mapa

map_obj
```

Criando outro mapa para visualizar os focos de calor como pontos individuais

```python
#criando outro mapa vazio
map = folium.Map(location = [-5.663417, -60.089654], zoom_start = 5)
map
```

Adicionando os pontos de focos de calor na região amazônica como pontos individuais e não como um mapa de calor, como realizado na visualização anterior

```python
amazonia = focos_amazonia  #necessário para atualizar o mapa quando necessário

for _, amazonia in amazonia.iterrows():
  folium.Marker(
      location = [amazonia['lat'], amazonia['lon']],
      ).add_to(map)

map
```

##DAQUI PRA FRENTE

Adicionando uma coordenada geográfica de interesse - Ponto de Referência para criar o buffer

```python
#ponto de referência
localizacao = focos_amazonia.loc[88516,'municipio']

latitude = -3.3767; longitude = -47.5882	
```

```python
m = folium.Map(location = [-5.663417, -60.089654], zoom_start = 5)

folium.Marker((latitude, longitude), popup = localizacao, 
              icon=folium.Icon(color="red", icon="info-sign")).add_to(m)
m
```

Transformar esse dataframe em um GeoDataframe

```python
home = pd.DataFrame([{'address': localizacao}])
home
```

Transformando agora em um GeoDataFrame

```python
home = gpd.tools.geocode(home['address'], Nominatim, user_agent='Isochrone calculator')
home
```

Isso é muito útil e podemos facilmente colocá-lo no mesmo mapa acima:

```python
map = folium.Map((latitude, longitude), zoom_start = 5)
folium.GeoJson(home, tooltip=folium.GeoJsonTooltip(['address'])).add_to(map)
map
```

Criando um buffer projetando na projeção GDA2020 MGA Zona 55 (Código ESG 7855) antes de reprojetar de volta a projeção original GCS WGS84

```python
buffer = home.to_crs(epsg=7855).buffer(5000).to_crs(epsg=4326)
```

```python
m = folium.Map((latitude, longitude), zoom_start = 10)
folium.GeoJson(buffer).add_to(m)
folium.GeoJson(home, tooltip=folium.GeoJsonTooltip(['address'])).add_to(m)
m
```

Adicionando todos os outros pontos novamente para verificar se algum deles está dentro do buffer, ou seja se algum foco de calor está dentro da zona de interesse determinado pelo buffer.

```python
amazonia = focos_amazonia  #necessário para atualizar o mapa quando necessário

for _, amazonia in amazonia.iterrows():
  folium.Marker(
      location = [amazonia['lat'], amazonia['lon']],
      ).add_to(m)

m
```
